# Centrality Measures and Categorical Analysis in Adolescent Social Networks

##  Centrality Measures and Social Outcomes: Assignment Summary

###  **Selected Dataset**
**Add Health High School Social Network Dataset**  
- Source: [UCI Network Data Repository – Add Health](http://moreno.ss.uci.edu/data.html#addhealth)  
- Description: This dataset contains friendship nomination data from a high school in Columbus, Ohio, collected as part of the National Longitudinal Study of Adolescent Health.

---

###  **Why This Dataset Works**
-  **Available on the web**: Publicly accessible from the University of California, Irvine website.
-  **Categorical variables** for each node:
  - **Gender** (Male, Female)
  - **Grade level** (7th–12th)
  - **Race**

---

###  **High-Level Plan to Load and Prepare the Data**
1. **Download files** from the UCI repository:
   - `.net` or `.gml` files contain the friendship network (nodes and edges).
   - Accompanying metadata files include categorical attributes.

2. **Load the network** using Python and NetworkX:
   - Use `nx.read_pajek()` or `nx.read_gml()` to import the network structure.

3. **Attach categorical metadata** to each node:
   - Parse the metadata file and map gender, grade, and race to node IDs.

4. **Calculate centrality measures**:
   - Degree Centrality
   - Eigenvector Centrality
   - (Optionally) Betweenness or Closeness Centrality

5. **Group nodes** by categorical variables and compare centrality scores across groups.

---

###  **Hypothetical Outcome to Predict**
> **Hypothesis:** Students with higher **degree centrality** (i.e., those who receive more friendship nominations) are more socially integrated and possibly more popular within their peer group.

We could predict:
- **Social integration level** based on **degree centrality**, and compare it:
  - Across **gender** (e.g., do female students have higher centrality on average?)
  - Across **grade level** (e.g., are seniors more central than freshmen?)

---

###  Conclusion
This dataset and approach provide a rich opportunity to explore how **network position** relates to **social standing**, using real-world categorical group comparisons. The full implementation of this analysis will be completed in **Project 1**.
