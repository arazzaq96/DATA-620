```python
import nltk

# Redownload punkt tokenizer from scratch
nltk.download('punkt', force=True)
```

```python
import re
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load corpus
from nltk.corpus import gutenberg
corpus = gutenberg.raw('austen-emma.txt')

# Step 2: Preprocess and tokenize using regex
words = re.findall(r'\b[a-z]+\b', corpus.lower())

# Step 3: Frequency analysis
word_freq = Counter(words)
total_words = sum(word_freq.values())
unique_words = len(word_freq)

print("Total unique words:", unique_words)
print("Total word count:", total_words)

# Step 4: How many words make up 50% of corpus?
running_total = 0
half_total = total_words / 2
unique_count = 0

for word, count in word_freq.most_common():
    running_total += count
    unique_count += 1
    if running_total >= half_total:
        break

print("Words that make up 50% of total:", unique_count)

# Step 5: Top 200 words
top_200 = word_freq.most_common(200)
top_words = [w for w, c in top_200]
top_counts = [c for w, c in top_200]

# Step 6: Plot
plt.figure(figsize=(12, 6))
plt.plot(top_counts)
plt.title("Top 200 Word Frequencies")
plt.xlabel("Word Rank")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Step 7: Zipf's Law comparison
ranks = np.arange(1, 201)
zipf = [top_counts[0] / r for r in ranks]

plt.figure(figsize=(12, 6))
plt.plot(ranks, top_counts, label='Observed')
plt.plot(ranks, zipf, linestyle='--', label="Zipf's Law Prediction")
plt.title("Zipf's Law vs Observed Frequencies")
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
