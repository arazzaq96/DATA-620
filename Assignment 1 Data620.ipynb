# Hello, Graph World

This notebook creates a simple graph structure using NetworkX and visualizes the Krackhardt Kite Graph.

## Tools Used
- Python
- NetworkX for graph creation
- Matplotlib for visualization
- Jupyter Notebook via Anaconda

## Graph Description
The Krackhardt Kite Graph is often used to illustrate social network concepts. It features a range of node degrees and is useful for demonstrating centrality measures.


```python
import networkx as nx
import matplotlib.pyplot as plt

```

```python

# Create an empty graph
G = nx.Graph()

# Add nodes
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
G.add_nodes_from(nodes)

# Add edges (Krackhardt kite graph edges)
edges = [
    ('A', 'B'), ('A', 'C'), ('A', 'D'),
    ('B', 'C'), ('B', 'D'),
    ('C', 'D'), ('D', 'E'), ('E', 'F'),
    ('F', 'G'), ('F', 'H'), ('G', 'H'),
    ('H', 'I'), ('I', 'J')
]
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=800, edge_color='gray')
plt.title("Krackhardt Kite Graph")
plt.show()
```
